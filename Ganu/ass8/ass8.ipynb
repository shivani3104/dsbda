```python
fig,axis=plt.subplots(2,2)
sns.histplot(ax=axis[0,0],data=df,x='Sex',hue='Survived')
sns.histplot(ax=axis[0,1],data=df,x='Pclass',hue='Survived')
sns.histplot(ax=axis[1,0],data=df,x='Age',hue='Survived')
sns.histplot(ax=axis[1,1],data=df,x='Fare',hue='Survived')
plt.show()
```

```python
fig, axis = plt.subplots(1,3)
sns.histplot(ax = axis[0], data = df, x='Sex', hue = 'Sex', multiple = 'dodge', shrink = 0.8)
sns.histplot(ax = axis[1], data = df, x='Pclass', hue = 'Pclass',multiple = 'dodge', shrink = 0.8)
sns.histplot(ax = axis[2], data = df, x='Survived', hue = 'Survived', multiple = 'dodge', shrink = 0.8)
plt.show()
```
