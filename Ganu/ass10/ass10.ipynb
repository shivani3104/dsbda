```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("iris.csv")
```

```python
df.columns=('sl','sw','pl','pw','species')
```

```python
print("The Basic information of dataset: " )
print('The information:\n', df.info)
print("The size of dataset:" , df.size)
print("The shape of dataset:" , df.shape)
print("The data types of dataset:\n" , df.dtypes)
print("The columns name :" , df.columns)
print("The first 5 rows:\n" , df.head())
print("The last 5 rows:\n" , df.tail())
print("Any 5 rows :\n" , df.sample(5))

```

```python
print("The statistical information of numerical information is :\n",df.describe())
```

```python
print("The Total null value in dataset:\n" , df.isnull().sum()) 
```

```python
fig,axis=plt.subplots(2,2)
sns.boxplot(ax=axis[0,0],data=df,y='sl')
sns.boxplot(ax=axis[0,1],data=df,y='sw')
sns.boxplot(ax=axis[1,0],data=df,y='pw')
sns.boxplot(ax=axis[1,1],data=df,y='pl')
plt.show()
```

```python
fig,axis=plt.subplots(2,2)
sns.boxplot(ax=axis[0,0],data=df,y='sl',hue='species')
sns.boxplot(ax=axis[0,1],data=df,y='sw',hue='species')
sns.boxplot(ax=axis[1,0],data=df,y='pw',hue='species')
sns.boxplot(ax=axis[1,1],data=df,y='pl',hue='species')
plt.show()
```

```python
fig,axis=plt.subplots(2,2)
sns.histplot(ax=axis[0,0],data=df,x='sl', multiple='dodge',shrink=0.8,kde=True)
sns.histplot(ax=axis[0,1],data=df,x='sw', multiple='dodge',shrink=0.8,kde=True)
sns.histplot(ax=axis[1,0],data=df,x='pl', multiple='dodge',shrink=0.8,kde=True)
sns.histplot(ax=axis[1,1],data=df,x='pw', multiple='dodge',shrink=0.8,kde=True)
plt.show()
```

```python
fig,axis=plt.subplots(2,2)
sns.histplot(ax=axis[0,0],data=df,x='sl',kde=True)
sns.histplot(ax=axis[0,1],data=df,x='sw',kde=True)
sns.histplot(ax=axis[1,0],data=df,x='pl',kde=True)
sns.histplot(ax=axis[1,1],data=df,x='pw',kde=True)
plt.show()
```
